```python
# unit test
D = np.random.randn(500).reshape((50,10))
D[0] = 2.
D[1] *= 3.
df_test = pd.DataFrame(D)
```

```python
def check_data(df,drop_cols=None):
    if drop_cols:
        row_var = df.drop(df.columns[drop_cols],axis=1).var(axis=1).as_matrix()
    else:
        row_var = df.var(axis=1).as_matrix()
    return df[row_var < .5]
```

```python
check_data(df_test)
```

```python
check_data(df,drop_cols=[0,1])
```
